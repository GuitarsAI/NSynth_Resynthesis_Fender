```python
# Importing Packages and Notebook Configurations
import os
import numpy as np
import matplotlib.pyplot as plt
from magenta.models.nsynth import utils
from magenta.models.nsynth.wavenet import fastgen
from IPython.display import Audio
import numpy as np, scipy, matplotlib.pyplot as plt, IPython.display as ipd
import librosa, librosa.display
%matplotlib inline
```

```python
# Importing audio file
filename = 'S_E2_NEC_FEN_STR_MX15665236_0_0.wav'
x, fs = librosa.load(filename, sr=16000)
fender_magenta=x/x.max()
sample_length_fender=fender_magenta.shape[0]
print('{} samples, {} seconds'.format(sample_length_fender, sample_length_fender / float(fs)))
```

```python
# Encoding
%time fender_encoding = fastgen.encode(fender_magenta, '../Magenta Test/wavenet-ckpt/model.ckpt-200000', sample_length_fender)
#http://download.magenta.tensorflow.org/models/nsynth/wavenet-ckpt.tar
```

```python
# Encodined shape
print(fender_encoding.shape)
np.save('fender_magenta' + '.npy', fender_encoding)
```

```python
# Plotting Waveforms
fig, axs = plt.subplots(2, 1, figsize=(20, 12))
axs[0].plot(fender_magenta);
axs[0].set_title('Fender - Audio Signal')
axs[0].grid(True)
axs[1].plot(fender_encoding[0]);
axs[1].set_title('Fender - Wavenet/Nsynth Encoding')
axs[1].grid(True);
```

```python
# Decode
%time fastgen.synthesize(fender_encoding, checkpoint_path='../Magenta Test/wavenet-ckpt/model.ckpt-200000', save_paths=['syn_' + filename], samples_per_save=sample_length_fender)
```

```python
# Load resynthesized audio
synthesis = utils.load_audio('syn_S_E2_NEC_FEN_STR_MX15665236_0_0.wav',
                             sample_length=sample_length_fender,
                             sr=fs)
```

```python
# Plot Original and Decoded Audio waveforms
plt.figure(figsize=(20,12))
plt.subplot(211)
librosa.display.waveplot(fender_magenta, sr=fs)
plt.grid(color='r', linestyle='--', linewidth=1, alpha=0.5)
plt.title('Fender Original')
plt.subplot(212)
librosa.display.waveplot(synthesis, sr=fs)
plt.title('Fender Wavenet/NSynth Decoded')
plt.grid(color='r', linestyle='--', linewidth=1, alpha=0.5)
plt.show()
```

```python
# Listen to audio
print('Fender Original')
ipd.display(ipd.Audio(fender_magenta, rate=fs))
print('Fender Re-synthesized')
ipd.display(ipd.Audio(synthesis, rate=fs))
```
